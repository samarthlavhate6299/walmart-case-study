# **Walmart Dataset**
The dataset contains information on customer purchases at Walmart, including demographic details, product categories, and purchase amounts.

- User_ID: Unique identifier for each customer.
- Product_ID: Unique identifier for each product.
- Gender: Gender of the customer, labeled as 'M' (Male) or 'F' (Female).
- Age: Age group of the customer, represented in ranges like '0-17', '26-35', '55+', etc.
- Occupation: Numerical code indicating the customer's occupation category.
- City_Category: Type of city the customer resides in, categorized as 'A', 'B', or 'C'. These labels likely represent city size or population, with 'A' being metropolitan, 'B' as mid-size cities, and 'C' as smaller or rural areas.
- Stay_In_Current_City_Years: The number of years the customer has lived in the current city, with values such as '1', '2', '3', or '4+'.
- Marital_Status: Indicates the customer's marital status, where 0 represents single and 1 represents married.
- Product_Category: Numerical code for the category of the purchased product, representing different product types.
- Purchase: The purchase amount spent by the customer in a single transaction, in currency units

## Understanding Distribution and Identifying Outliers

```python
Q1 = df['Purchase'].quantile(0.25)
Q3 = df['Purchase'].quantile(0.75)
IQR = Q3 - Q1
Purchase_lower_bound = Q1 - 1.5 * IQR
Purchase_upper_bound = Q3 + 1.5 * IQR
print(f"Income Upper Bound: {Purchase_upper_bound}, Income Lower Bound: {Purchase_lower_bound}")
is_outlier_upper = Purchase_upper_bound < df['Purchase'].max()
is_outlier_lower = Purchase_lower_bound > df['Purchase'].min()
print(f"Is the value exceeding the upper bound limit? {is_outlier_upper}")
print(f"Is the value falling below the lower bound limit? {is_outlier_lower}.")
print('---'*30)
```

```python
clean_Purchase_data = df[(df['Purchase'] >= Purchase_lower_bound) & (df['Purchase'] <=Purchase_upper_bound)]
```

## Confidence intervals and distribution of the mean of the expenses by female and male customers (10 Points)

```python
def confidence_interval(data,ci):
    l_ci = (100-ci)/2
    u_ci = (100+ci)/2
    interval = np.percentile(data,[l_ci,u_ci]).round(0)

    return interval
```

```python

def plot(ci):
    fig = plt.figure(figsize = (15,8))
    gs = fig.add_gridspec(2,2)
    df_male = df.loc[df['Gender'] == 'M','Purchase']
    df_female = df.loc[df['Gender'] == 'F','Purchase']
    sample_sizes = [(100,0,0),(1000,0,1),(3000,1,0),(10000,1,1)]
    bootstrap_samples = 2000

    male_samples = {}
    female_samples = {}

    for i,x,y in sample_sizes:
        male_means = []
        female_means = []

        for j in range(bootstrap_samples):
            male_bootstrapped_samples = np.random.choice(df_male,size = i)
            female_bootstrapped_samples = np.random.choice(df_female,size = i)

            male_sample_mean = np.mean(male_bootstrapped_samples)
            female_sample_mean = np.mean(female_bootstrapped_samples)

            male_means.append(male_sample_mean)
            female_means.append(female_sample_mean)

        male_samples[f'{ci}%_{i}'] = male_means
        female_samples[f'{ci}%_{i}'] = female_means

        temp_df = pd.DataFrame(data = {'male_means':male_means,'female_means':female_means})
        ax = fig.add_subplot(gs[x,y])

        sns.kdeplot(data = temp_df,x = 'male_means' ,fill = True, alpha = 1,ax = ax,label = 'Male')
        sns.kdeplot(data = temp_df,x = 'female_means' ,fill = True, alpha = 1,ax = ax,label = 'Female')

        m_range = confidence_interval(male_means,ci)
        f_range = confidence_interval(female_means,ci)


        for k in m_range:
            ax.axvline(x = k,ymax = 0.9,linestyle = '--')

        for k in f_range:
            ax.axvline(x = k,ymax = 0.9, linestyle = '--')


        for s in ['top','left','right']:
            ax.spines[s].set_visible(False)

        ax.set_yticks([])
        ax.set_ylabel('')
        ax.set_xlabel('')

        ax.set_title(f'CLT Curve for Sample Size = {i}')

        plt.legend()

    fig.suptitle(f'{ci}% Confidence Interval')

    plt.show()

    return male_samples,female_samples
```

```python
m_samp_90,f_samp_90 = plot(90)
```

#### **Observations:**

**Specific Observations**:

- For sample size 100: The distributions for both males and females are relatively wide, with some overlap. This suggests that with smaller sample sizes, there is more uncertainty in estimating the population mean.
- For sample size 1000: The distributions become narrower and more concentrated around the population mean. The overlap between the male and female distributions is reduced, suggesting that we can differentiate between the two genders with more confidence.
- For sample size 3000 and 10000: The distributions become even more concentrated, with minimal overlap between the genders. This indicates high precision in estimating the population mean for both genders.

**Recommendations**:

- To obtain more accurate and reliable estimates of the population mean, it is advisable to use larger sample sizes.
- The appropriate sample size depends on the desired level of precision and the variability in the population.
- Consider using statistical tests to determine if the observed differences in means between genders are statistically significant.

**Overall**:

- As the sample size increases, the distribution of the sample means becomes more concentrated around the true population mean.
- The confidence intervals become narrower with larger sample sizes, indicating increased precision in estimating the population mean.

## **Results when the same activity is performed for Married vs Unmarried**

```python
def analyze_martial_group_spending_with_clt(df, confidence_level=0.95, num_bootstraps=1000):

    grouped_data = df.groupby('converted_Martial_Status')['Purchase'].agg(['mean', 'std']).reset_index()

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
    axes = axes.flatten()

    for i, (index, row) in enumerate(grouped_data.iterrows()):
        marital_status = row['converted_Martial_Status']
        sample_size = df[df['converted_Martial_Status'] == marital_status]['Purchase'].shape[0]
        sample_mean = row['mean']
        sample_std = row['std']

        bootstrap_means = []
        for _ in range(num_bootstraps):
            bootstrap_sample = np.random.choice(df[df['converted_Martial_Status'] == marital_status]['Purchase'], size=sample_size, replace=True)
            bootstrap_means.append(bootstrap_sample.mean())
        std_error = sample_std / np.sqrt(sample_size)
        z_critical = norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = z_critical * std_error
        confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
        print(f'{marital_status} average spent value and Confidence Interval:')
        print(f'{marital_status} customer average spent amount: {sample_mean}')
        print(f'Confidence Interval: {confidence_interval}')
        print()
        sns.kdeplot(bootstrap_means, ax=axes[i])
        axes[i].axvline(x=confidence_interval[0], color='red', linestyle='--', label='Confidence Interval')
        axes[i].axvline(x=confidence_interval[1], color='red', linestyle='--')

        axes[i].set_title(f'Distribution of Mean Purchase Amounts for {marital_status}')
        axes[i].set_xlabel('Mean Purchase Amount')
        axes[i].set_ylabel('Density')

    plt.tight_layout()
    plt.show()

analyze_martial_group_spending_with_clt(df)
```

#### **Observations:**
**Confidence Interval Overlap**:

- The confidence intervals for married and unmarried customers' average spending overlap. This suggests that there might not be a statistically significant difference between the average spending of married and unmarried customers.

**Implications for Walmart**:

- No Need for Separate Strategies: Since there's no significant difference in average spending, Walmart may not need to implement separate marketing or product strategies for married and unmarried customers.
- Focus on Other Factors: Walmart should consider other factors, such as age, income, or lifestyle preferences, to identify potential customer segments and tailor marketing efforts accordingly.
